```python
# Count the rows before removing duplicates
before_rows = len(df)

# drop_duplicates() removes repeated rows from the DataFrame
# inplace=True changes df directly instead of creating a new copy
df.drop_duplicates(inplace=True)

# Count the rows after removing duplicates
after_rows = len(df)

# Show how many duplicate rows were removed
print(f"Removed {before_rows - after_rows} duplicate rows")

# Confirm the remaining duplicate count
print(df.duplicated().sum())
```

```python
df
```

```python
# Find rows that look unusual or broken in common columns
# We treat a row as abnormal if:
# - year is not a 4-digit number
# - Price is missing or says "Ask For Price"
# - kms_driven does not contain a number followed by "kms"
# - fuel_type is missing

year_ok = df['year'].astype(str).str.fullmatch(r'\d{4}').fillna(False)

abnormal_rows = df[
    (~year_ok)
    | (df['Price'].astype(str).eq('Ask For Price'))
    | (df['Price'].isna())
    | (~df['kms_driven'].astype(str).str.contains(r'\d+\s*kms', na=False))
    | (df['fuel_type'].isna())
]

# Show the suspicious rows so you can inspect them
abnormal_rows
```

```python
# Clean columns that should contain numbers
# If a value is not a number, pandas changes it to NaN

df['year'] = pd.to_numeric(df['year'], errors='coerce')

# Remove rows where year is not a valid number
before_rows = len(df)
df = df.dropna(subset=['year']).copy()
df['year'] = df['year'].astype(int)
after_rows = len(df)
print(f"Removed {before_rows - after_rows} rows with bad year values")

# Clean kms_driven by removing the word 'kms' and commas first
df['kms_driven'] = df['kms_driven'].astype(str).str.replace(' kms', '', regex=False)
df['kms_driven'] = df['kms_driven'].str.replace(',', '', regex=False)
df['kms_driven'] = pd.to_numeric(df['kms_driven'], errors='coerce')

# Remove rows where kms_driven is not a valid number
before_rows = len(df)
df = df.dropna(subset=['kms_driven']).copy()
df['kms_driven'] = df['kms_driven'].astype(int)
after_rows = len(df)
print(f"Removed {before_rows - after_rows} rows with bad kms_driven values")

# Clean Price by removing commas
df['Price'] = df['Price'].astype(str).str.replace(',', '', regex=False)

# Turn Price into numbers too
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Show the cleaned dataset
df
```
